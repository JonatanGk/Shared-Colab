````
from transformers import pipeline
````

## 1. Create the pipeline

````
model_path = "JonatanGk/roberta-base-ca-finetuned-cyberbullying-catalan"
bullying_analysis = pipeline("text-classification", model=model_path, tokenizer=model_path)
````

## 2. Make analysis

````
bullying_analysis(
    "Des que et vaig veure m'en vaig enamorar de tu."
    )
````

````
bullying_analysis(
    "Ets tan lletja que quan eres petita et donavan de menjar per sota la porta."
    )
````

````
bullying_analysis(
    "Ets més petit que un barrufet!!"
    )
````
